```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3DRealme&otracker=nmenu_sub_Electronics_0_Realme"
```

```python
x = requests.get(url)
x
```

```python
soup = BeautifulSoup(x.content,"html.parser")
print(soup.prettify())
```

```python
sushil = soup.find("div",{"class" :"_4rR01T"})
sushil.text
```

```python
sushil = soup.find_all("div",{"class" :"_4rR01T"})
for i in sushil:
    print(i.text)
```

```python
mob_name  = []

sushil = soup.find_all("div",{"class" :"_4rR01T"})
for i in sushil:
    mob_name.append(i.text)
    
    
    
```
